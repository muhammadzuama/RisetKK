```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
import time

# ==================================================
# War Strategy Optimization (Binary Version)
# ==================================================
def run_wso_feature_selection(X_train, y_train, n_agents=30, iters=40):
	dim = X_train.shape[1]
	start_time = time.time()

	# ---------------------------------------------
	# Objective function (same as PSO/WOA)
	# ---------------------------------------------
	def fitness(mask):
		selected = np.where(mask == 1)[0]
		if len(selected) < 2:
			return 1.0
		X_sel = X_train[:, selected]
		acc = cross_val_score(KNeighborsClassifier(n_neighbors=5),
							  X_sel, y_train, cv=5).mean()
		return 1 - acc
	
	# ---------------------------------------------
	# Initialize population
	# ---------------------------------------------
	pop = np.random.randint(0, 2, (n_agents, dim))
	fitness_values = np.array([fitness(ind) for ind in pop])

	best_idx = np.argmin(fitness_values)
	best_sol = pop[best_idx].copy()
	best_fit = fitness_values[best_idx]

	# ---------------------------------------------
	# WSO main loop
	# ---------------------------------------------
	for it in range(iters):
		for i in range(n_agents):

			# choose a random strategy
			strategy = np.random.choice(["attack", "defend", "ambush", "retreat"])

			new = pop[i].copy()

			if strategy == "attack":
				# move towards the best solution
				new = np.where(np.random.rand(dim) < 0.7, best_sol, new)

			elif strategy == "defend":
				# stay stable with small mutation
				mutation = (np.random.rand(dim) < 0.1).astype(int)
				new = np.logical_xor(new, mutation).astype(int)

			elif strategy == "ambush":
				# random aggressive changes
				rand = np.random.randint(0, 2, dim)
				new = np.where(np.random.rand(dim) < 0.3, rand, new)

			elif strategy == "retreat":
				# diversify: flip bits with low probability
				new = np.where(np.random.rand(dim) < 0.05, 1-new, new)

			# evaluate
			new_fit = fitness(new)

			# greedy selection
			if new_fit < fitness_values[i]:
				pop[i] = new
				fitness_values[i] = new_fit

				# update global best
				if new_fit < best_fit:
					best_sol = new.copy()
					best_fit = new_fit
		
		print(f"Iteration {it+1}/{iters} → best fitness = {best_fit:.4f}")
	
	end_time = time.time()
	print(f"\nWSO selected {np.sum(best_sol)} features out of {dim}")
	print(f"⏱ Optimization time: {end_time - start_time:.2f} seconds")
	return np.where(best_sol == 1)[0]
```

```python
# %%
import time
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

def evaluate_models(models, X_train, X_test, y_train, y_test, average='weighted', plot_confusion=True):
	"""
	Fungsi untuk mengevaluasi beberapa classifier sekaligus.
	
	Parameters:
	- models        : list of tuples [('nama_model', model_instance), ...]
	- X_train, X_test, y_train, y_test
	- average       : average method untuk multi-class (default='weighted')
	- plot_confusion: True/False tampilkan confusion matrix
	
	Returns:
	- DataFrame berisi metric untuk semua model
	"""
	results_list = []

	for name, model in models:
		print(f"\n========== Evaluating: {name} ==========")
		
		# Training
		start_train = time.time()
		model.fit(X_train, y_train)
		end_train = time.time()
		
		# Prediction
		start_test = time.time()
		y_pred = model.predict(X_test)
		end_test = time.time()
		
		# Metric
		acc = accuracy_score(y_test, y_pred)
		prec = precision_score(y_test, y_pred, average=average)
		rec = recall_score(y_test, y_pred, average=average)
		f1 = f1_score(y_test, y_pred, average=average)
		
		print(f"Accuracy : {acc:.4f}")
		print(f"Precision: {prec:.4f}")
		print(f"Recall   : {rec:.4f}")
		print(f"F1-Score : {f1:.4f}")
		print(f"⏱ Training time: {end_train - start_train:.2f} sec")
		print(f"⏱ Testing time : {end_test - start_test:.2f} sec")
		
		# Classification report + confusion matrix
		if plot_confusion:
			print("\nClassification Report:\n")
			print(classification_report(y_test, y_pred, digits=4))
			cm = confusion_matrix(y_test, y_pred)
			plt.figure(figsize=(5,4))
			sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
			plt.title(f'Confusion Matrix: {name}')
			plt.xlabel('Predicted')
			plt.ylabel('Actual')
			plt.show()
		
		# Simpan hasil
		results_list.append({
			'Model': name,
			'Accuracy': acc,
			'Precision': prec,
			'Recall': rec,
			'F1-Score': f1,
			'Train_Time(s)': end_train - start_train,
			'Test_Time(s)': end_test - start_test
		})
	
	return pd.DataFrame(results_list)
```

```python
selected_80 = run_wso_feature_selection(X_train_scaled_80, y_train, n_agents=30, iters=40)
X_train_final_80 = X_train_scaled_80[:, selected_80]
X_test_final_80 = X_test_scaled_80[:, selected_80]
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

models = [
	('Random Forest', RandomForestClassifier(n_estimators=100, random_state=42)),
	('Gradient Boosting', GradientBoostingClassifier(n_estimators=200, learning_rate=0.1, max_depth=3, random_state=42)),
	('Naive Bayes', GaussianNB()),
	('SVM', SVC(kernel='rbf', C=1.0, gamma='scale', random_state=42))
]

results_df = evaluate_models(models, X_train_final_80, X_test_final_80, y_train, y_test)
print("\n✅ Summary of all models:\n")
print(results_df)
```
