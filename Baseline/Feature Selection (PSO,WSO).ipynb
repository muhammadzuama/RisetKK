## Feature Selection Function

### PSO Function

```python
# ==========================
# Fungsi PSO Feature Selection
# ==========================
def run_pso_feature_selection(X_train, y_train, n_particles, iters=40):
    """
    Binary PSO untuk seleksi fitur dengan KNN
    """
    from pyswarms.discrete import BinaryPSO
    from sklearn.neighbors import KNeighborsClassifier
    from sklearn.model_selection import cross_val_score
    import numpy as np
    import time

    def pso_objective(mask):
        scores = []
        for i in range(mask.shape[0]):
            selected = np.where(mask[i] == 1)[0]
            if len(selected) < 2:
                scores.append(1.0)
                continue
            X_sel = X_train[:, selected]
            acc = cross_val_score(KNeighborsClassifier(n_neighbors=5),
                                  X_sel, y_train, cv=5).mean()
            scores.append(1 - acc)
        return np.array(scores)

    optimizer = BinaryPSO(
        n_particles=n_particles,
        dimensions=X_train.shape[1],
        options={'c1': 1.5, 'c2': 1.5, 'w': 0.7, 'k': 10, 'p': 2}
    )

    start_time = time.time()
    best_cost, best_mask = optimizer.optimize(pso_objective, iters=iters)
    end_time = time.time()

    selected_features = np.where(best_mask == 1)[0]

    print(f"✅ PSO selected {len(selected_features)} features out of {X_train.shape[1]}")
    print(f"⏱ Optimization time: {end_time - start_time:.2f} seconds")

    return selected_features
```

### WSO Function

```python
# ==========================
# Fungsi WOA Feature Selection
# ==========================
def run_woa_feature_selection(X_train, y_train, n_whales, iters=40):
    """
    Binary Whale Optimization Algorithm (WOA) untuk seleksi fitur dengan KNN
    """
    import numpy as np
    from sklearn.neighbors import KNeighborsClassifier
    from sklearn.model_selection import cross_val_score
    import time

    # --- Objective function: minimize (1 - accuracy) ---
    def woa_objective(mask):
        scores = []
        for i in range(mask.shape[0]):
            selected = np.where(mask[i] == 1)[0]
            if len(selected) < 2:
                scores.append(1.0)
                continue
            X_sel = X_train[:, selected]
            acc = cross_val_score(KNeighborsClassifier(n_neighbors=5),
                                  X_sel, y_train, cv=5).mean()
            scores.append(1 - acc)
        return np.array(scores)

    # --- Initialization ---
    n_features = X_train.shape[1]
    whales = np.random.randint(0, 2, (n_whales, n_features))
    best_whale = None
    best_score = np.inf

    start_time = time.time()

    for t in range(iters):
        # Evaluate each whale
        fitness = woa_objective(whales)
        # Update best whale
        min_idx = np.argmin(fitness)
        if fitness[min_idx] < best_score:
            best_score = fitness[min_idx]
            best_whale = whales[min_idx].copy()

        # Linearly decrease parameter 'a' from 2 to 0
        a = 2 - t * (2 / iters)

        for i in range(n_whales):
            r1, r2 = np.random.rand(), np.random.rand()
            A = 2 * a * r1 - a
            C = 2 * r2
            p = np.random.rand()

            if p < 0.5:
                if abs(A) >= 1:
                    rand_whale = whales[np.random.randint(n_whales)]
                    D = abs(C * rand_whale - whales[i])
                    new_position = rand_whale - A * D
                else:
                    D = abs(C * best_whale - whales[i])
                    new_position = best_whale - A * D
            else:
                D = abs(best_whale - whales[i])
                b = 1  # spiral constant
                l = np.random.uniform(-1, 1)
                new_position = D * np.exp(b * l) * np.cos(2 * np.pi * l) + best_whale

            # --- Sigmoid transfer function for binary conversion ---
            S = 1 / (1 + np.exp(-new_position))
            whales[i] = np.where(np.random.rand(n_features) < S, 1, 0)

        print(f"Iter {t+1}/{iters} | Best fitness: {best_score:.4f}")

    end_time = time.time()

    selected_features = np.where(best_whale == 1)[0]

    print(f"✅ WOA selected {len(selected_features)} features out of {n_features}")
    print(f"⏱ Optimization time: {end_time - start_time:.2f} seconds")

    return selected_features
```

## Model Evaluations

### Define Function

```python
# %%
import time
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

def evaluate_models(models, X_train, X_test, y_train, y_test, average='weighted', plot_confusion=True):
    """
    Fungsi untuk mengevaluasi beberapa classifier sekaligus.
    
    Parameters:
    - models        : list of tuples [('nama_model', model_instance), ...]
    - X_train, X_test, y_train, y_test
    - average       : average method untuk multi-class (default='weighted')
    - plot_confusion: True/False tampilkan confusion matrix
    
    Returns:
    - DataFrame berisi metric untuk semua model
    """
    results_list = []

    for name, model in models:
        print(f"\n========== Evaluating: {name} ==========")
        
        # Training
        start_train = time.time()
        model.fit(X_train, y_train)
        end_train = time.time()
        
        # Prediction
        start_test = time.time()
        y_pred = model.predict(X_test)
        end_test = time.time()
        
        # Metric
        acc = accuracy_score(y_test, y_pred)
        prec = precision_score(y_test, y_pred, average=average)
        rec = recall_score(y_test, y_pred, average=average)
        f1 = f1_score(y_test, y_pred, average=average)
        
        print(f"Accuracy : {acc:.4f}")
        print(f"Precision: {prec:.4f}")
        print(f"Recall   : {rec:.4f}")
        print(f"F1-Score : {f1:.4f}")
        print(f"⏱ Training time: {end_train - start_train:.2f} sec")
        print(f"⏱ Testing time : {end_test - start_test:.2f} sec")
        
        # Classification report + confusion matrix
        if plot_confusion:
            print("\nClassification Report:\n")
            print(classification_report(y_test, y_pred, digits=4))
            cm = confusion_matrix(y_test, y_pred)
            plt.figure(figsize=(5,4))
            sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
            plt.title(f'Confusion Matrix: {name}')
            plt.xlabel('Predicted')
            plt.ylabel('Actual')
            plt.show()
        
        # Simpan hasil
        results_list.append({
            'Model': name,
            'Accuracy': acc,
            'Precision': prec,
            'Recall': rec,
            'F1-Score': f1,
            'Train_Time(s)': end_train - start_train,
            'Test_Time(s)': end_test - start_test
        })
    
    return pd.DataFrame(results_list)
```
