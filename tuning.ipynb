```python
# Pisahkan fitur dan label
X = data.drop(columns=['Diagnosis'])   # ganti 'Diagnosis' dengan nama kolom target kamu
y = data['Diagnosis']
```

```python
from sklearn.model_selection import train_test_split
# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

print("training_data : ", X_train.shape)
print("test_data : ", X_test.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Normalisasi MinMax
scaler = MinMaxScaler()

# Normalisasi 80:20
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)


print(f"Data Train after normalisation = {X_train_scaled.shape}")
print(f"Data Test after normalisation = {X_test_scaled.shape}")
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import f1_score, roc_auc_score, classification_report, roc_curve, auc
import numpy as np

# ===========================
# 🔍 1. Definisi Model Dasar
# ===========================
gb_model = GradientBoostingClassifier(random_state=42)

# ===========================
# 🎯 2. Grid Hyperparameter
# ===========================
param_grid = {
    "n_estimators": [100, 200, 300, 400],
    "learning_rate": [0.01, 0.05, 0.1],
    "max_depth": [3, 4, 5, 6],
    "subsample": [0.7, 0.8, 1.0]
}

# ===========================
# ⚙️ 3. GridSearchCV
# F1-score = metric utama
# AUC = tambahan evaluasi
# ===========================
grid_search = GridSearchCV(
    estimator=gb_model,
    param_grid=param_grid,
    scoring="f1_macro",         # bisa ganti f1_weighted kalau imbalance berat
    cv=5,
    n_jobs=-1,
    verbose=2
)

# ===========================
# 🚀 4. Fit ke data train
# ===========================
grid_search.fit(X_train_scaled, y_train)

# ===========================
# 🏆 5. Tampilkan hasil terbaik
# ===========================
print("Best Parameters:")
print(grid_search.best_params_)

print("\nBest CV F1 Score:")
print(grid_search.best_score_)

# ===========================
# 🔥 6. Evaluasi di Data Test
# ===========================
best_model = grid_search.best_estimator_

y_pred = best_model.predict(X_test_scaled)
y_proba = best_model.predict_proba(X_test_scaled)[:, 1]  # probabilities untuk AUC

# --- F1-score test ---
f1_test = f1_score(y_test, y_pred, average="macro")

# --- AUC-ROC test ---
try:
    auc_test = roc_auc_score(y_test, y_proba)
except:
    auc_test = "AUC tidak dapat dihitung (mungkin multiclass tanpa probability)"

print("\n=== Test Evaluation ===")
print(f"F1-score (macro): {f1_test:.4f}")
print(f"AUC-ROC: {auc_test}")

print("\n=== Classification Report ===")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import f1_score, roc_auc_score, classification_report
import numpy as np

# ===========================
# 🔍 1. Definisi Model Dasar
# ===========================
rf_model = RandomForestClassifier(random_state=42)

# ===========================
# 🎯 2. Grid Hyperparameter
# ===========================
param_grid = {
    "n_estimators": [100, 200, 300, 500],
    "max_depth": [None, 5, 10, 20],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4],
    "max_features": ["sqrt", "log2", None]
}

# ===========================
# ⚙️ 3. GridSearchCV
# ===========================
grid_search = GridSearchCV(
    estimator=rf_model,
    param_grid=param_grid,
    scoring="f1_macro",     # pakai f1 sebagai metric utama
    cv=5,
    n_jobs=-1,
    verbose=2
)

# ===========================
# 🚀 4. Training (CV)
# ===========================
grid_search.fit(X_train_scaled, y_train)

# ===========================
# 🏆 5. Hasil terbaik
# ===========================
print("Best Parameters:")
print(grid_search.best_params_)

print("\nBest CV F1 Score:")
print(grid_search.best_score_)

# ===========================
# 🔥 6. Evaluasi di Data Test
# ===========================
best_rf = grid_search.best_estimator_

y_pred = best_rf.predict(X_test_scaled)
y_proba = best_rf.predict_proba(X_test_scaled)[:, 1]

# --- F1-score test ---
f1_test = f1_score(y_test, y_pred, average="macro")

# --- AUC-ROC test ---
try:
    auc_test = roc_auc_score(y_test, y_proba)
except:
    auc_test = "AUC tidak dapat dihitung (mungkin multiclass)"

print("\n=== Test Evaluation ===")
print(f"F1-score (macro): {f1_test:.4f}")
print(f"AUC-ROC: {auc_test}")

print("\n=== Classification Report ===")
print(classification_report(y_test, y_pred))
```
