```python
# Pisahkan fitur dan label
X = data.drop(columns=['Diagnosis'])   # ganti 'Diagnosis' dengan nama kolom target kamu
y = data['Diagnosis']
```

```python
from sklearn.model_selection import train_test_split
# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

print("training_data : ", X_train.shape)
print("test_data : ", X_test.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Normalisasi MinMax
scaler = MinMaxScaler()

# Normalisasi 80:20
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)


print(f"Data Train after normalisation = {X_train_scaled.shape}")
print(f"Data Test after normalisation = {X_test_scaled.shape}")
```

```python
# ==========================
# Fungsi PSO Feature Selection (Stable Version)
# ==========================
def run_pso_feature_selection(X_train, y_train, n_particles, iters=60):
    """
    Binary PSO untuk seleksi fitur dengan KNN (versi stabil)
    """
    import numpy as np
    import random
    import time
    from pyswarms.discrete import BinaryPSO
    from sklearn.neighbors import KNeighborsClassifier
    from sklearn.model_selection import cross_val_score

    # === FIX 1: Stabilkan random ===
    # np.random.seed(42)
    # random.seed(42)

    def pso_objective(mask):
        scores = []
        for i in range(mask.shape[0]):
            selected = np.where(mask[i] == 1)[0]
            if len(selected) < 2:
                scores.append(1.0)
                continue
            X_sel = X_train[:, selected]

            acc = cross_val_score(
                KNeighborsClassifier(n_neighbors=5),
                X_sel, y_train,
                cv=5,
                n_jobs=-1
            ).mean()

            # minimize → 1 - acc
            scores.append(1 - acc)
        return np.array(scores)

    # === FIX 2: parameter PSO dibuat stabil ===
    options = {
        'c1': 1.3,
        'c2': 1.3,
        'w': 0.8,      # inertia agak besar untuk stabil
        'k': 15,       # neighborhood lebih besar
        'p': 2
    }

    optimizer = BinaryPSO(
        n_particles=n_particles,
        dimensions=X_train.shape[1],
        options=options
    )

    start = time.time()
    # === FIX 3: tambahkan tolerance agar berhenti jika sudah stabil ===
    best_cost, best_mask = optimizer.optimize(
        pso_objective,
        iters=iters
    )
    end = time.time()

    selected_features = np.where(best_mask == 1)[0]

    print(f"Selected Features : {len(selected_features)}")
    print(f"Time              : {end - start:.2f} sec")

    return selected_features
```

```python
# Jalankan PSO
selected = run_pso_feature_selection(X_train_scaled, y_train, n_particles=30)
X_train_final = X_train_scaled[:, selected]
X_test_final = X_test_scaled[:, selected]
```

```python
import time
import numpy as np
from sklearn.metrics import (
    recall_score, f1_score, confusion_matrix, classification_report, roc_auc_score
)
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

def evaluate_models(models, X_train, X_test, y_train, y_test, average='weighted', 
                    plot_confusion=True, save_csv=True, csv_path="hasil_evaluasi.csv"):
    """
    Evaluasi beberapa model dan simpan hasilnya ke CSV.
    """
    results_list = []

    for name, model in models:
        print(f"\n========== Evaluating: {name} ==========")
        
        # Training
        start_train = time.time()
        model.fit(X_train, y_train)
        end_train = time.time()
        
        # Prediction
        start_test = time.time()
        y_pred = model.predict(X_test)
        end_test = time.time()
        
        # Probability untuk AUCROC
        try:
            y_proba = model.predict_proba(X_test)
            proba_available = True
        except:
            y_proba = None
            proba_available = False
        
        # Metrics
        f1 = f1_score(y_test, y_pred, average=average)
        rec = recall_score(y_test, y_pred, average=average)

        # AUC-ROC
        if proba_available:
            if len(np.unique(y_test)) > 2:
                auc = roc_auc_score(y_test, y_proba, multi_class='ovr', average='weighted')
            else:
                auc = roc_auc_score(y_test, y_proba[:, 1])
        else:
            auc = np.nan
        
        print(f"F1-Score : {f1:.4f}")
        print(f"Recall   : {rec:.4f}")
        print(f"AUC-ROC  : {auc:.4f}")
        print(f"⏱ Training time: {end_train - start_train:.2f} sec")
        print(f"⏱ Testing time : {end_test - start_test:.2f} sec")
        
        # Classification report + Confusion matrix
        if plot_confusion:
            print("\nClassification Report:\n")
            print(classification_report(y_test, y_pred, digits=4))
            
            cm = confusion_matrix(y_test, y_pred)
            plt.figure(figsize=(5,4))
            sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
            plt.title(f'Confusion Matrix: {name}')
            plt.xlabel('Predicted')
            plt.ylabel('Actual')
            plt.show()
        
        # Save results into list
        results_list.append({
            'Model': name,
            'F1-Score': f1,
            'Recall': rec,
            'AUC-ROC': auc,
            'Train_Time(s)': end_train - start_train,
            'Test_Time(s)': end_test - start_test
        })
    
    # Convert ke DataFrame
    df_results = pd.DataFrame(results_list)

    # Simpan ke CSV
    if save_csv:
        df_results.to_csv(csv_path, index=False)
        print(f"\n📁 Hasil evaluasi berhasil disimpan ke: {csv_path}")

    return df_results
```

```python
# %%
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

# ===============================
# Daftar model untuk evaluasi
# ===============================
models = [
    ('Random Forest', RandomForestClassifier(random_state=42)),
    ('Gradient Boosting', GradientBoostingClassifier(random_state=42)),  # 🔹 default
    ('Naive Bayes', GaussianNB()),
    ('SVM (RBF)', SVC(random_state=42)),  # 🔹 default kernel = 'rbf'
]

# ===============================
# Evaluasi semua model
# ===============================
results_df = evaluate_models(
    models,
    X_train_final,
    X_test_final,
    y_train,   
    y_test
)

print("\n✅ Summary of all models:\n")
print(results_df)

```
